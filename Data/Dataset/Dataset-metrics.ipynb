```python
import json
import pandas as pd
```

```python
cities = ['Bristol', 'Leeds', 'London', 'Manchester', 'West-Midlands']
full_data = []
max_block_id=120

for city in cities:
    file = open(city + '.geojson',)
    compactnes_file = open('../../Evaluator/Compatcness-Evaluation/' + city + '.compact')
    
    geojson_content = json.load(file)
    file.close()
    
    compactness_content = [float(line.split('\n')[0]) for line in compactnes_file.readlines()]
    compactnes_file.close()
    
    for index, block in enumerate(geojson_content['features']):
        block_id = int(block['properties']['ID'])
        
        if block_id >= max_block_id:
            continue
        
        name = block['properties']['NAME']
        population = block['properties']['POPULATION']
        area = block['properties']['AREA']
        density = block['properties']['DENSITY']

        data = {
            'id': block_id,
            'city': city,
            'name': name,
            'density': density,
            'population': population,
            'area': area,
            'compactness': compactness_content[index]
        }
        
        full_data.append(data)
        
```

```python
df = pd.DataFrame(full_data)
df.head(123)
```

```python
df_by_city = df.groupby('city').agg({'density': ['describe'],'population': ['describe'], 'area': ['describe'], 'compactness': ['describe']})
df_by_city.head()
```

```python
df_transpose = df_by_city.transpose()
df_transpose
```
